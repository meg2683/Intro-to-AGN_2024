#### 1. Plot spectrum of PNe for different abundances and discuss why we see differences in spectra.

Input file to cloudy.exe for simulating planetary abundances:

And for solar abundances:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pne=pd.read_csv('pn.con', sep='\t')
pne.head()
```

```python
solar=pd.read_csv('solar.con', sep='\t')
solar.head()
```

```python
fig = plt.figure(figsize=(10,8))

plt.plot(pne['#Cont  nu'], pne['net trans'], label='PNe abundances', color='blue')
plt.plot(solar['#Cont  nu'], solar['net trans'], label='Solar abundances', color = 'red',lw=0.75)

plt.title('Spectral Energy Distribution')
plt.xlabel('Wavelength [$\mu$m]')
plt.ylabel('Flux [erg cm$^{-2}$ s$^{-1}$]')

plt.xscale('log')
plt.yscale('log')

plt.xlim(0.1, 1200)
plt.ylim(1e33, 2e40)

plt.legend()
fig.savefig('PNe_spectra.png')
```

- The most striking feature in the planetary nebula spectra is the large emission feature centered at wavelength 30$\mu \text{m}$. This is produced by dust in the nebula that is heated by photons from the central star. 
- The recombination continua produces the cliff-like features in the optical and near infrared. 
- The emission lines are produced by both recombination and collisional excitation.

#### 3. Plot the temperature of the PNe from the overview file, i.e. the distance from the center (first column) vs. gas temperature (second column)

```python
pne_ovr=pd.read_csv('pn.ovr', sep='\t')
pne_ovr.head()
```

```python
solar_ovr=pd.read_csv('solar.ovr', sep='\t')
solar_ovr.head()
```

```python
fig = plt.figure(figsize=(10,8))


plt.plot(pne_ovr['#depth'], pne_ovr['Te'], label='PNe abundances')
plt.plot(solar_ovr['#depth'], solar_ovr['Te'], label='Solar abundances')

plt.title('Temperatures of Planetary Nebulae')
plt.xlabel('Distance from center [cm]')
plt.ylabel('Temperature [K]')

plt.xscale('log')
plt.legend()
fig.savefig('PNe_temp.png')
```
